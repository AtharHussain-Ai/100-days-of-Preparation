1 : What is Simple Linear Regression ?

Ans: Simple Linear Regression is a method use for predect a depended variable Y , based on independent variable X it tries to fit the best fit straight line through data points.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
X = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
```

```python
X
```

```python
Y = np.array([1.5, 3.1, 4.2, 5.8, 7.5])
```

```python
Y
```

```python
model = LinearRegression()
```

```python
model.fit(X,Y)
```

```python
m = model.coef_[0]
c = model.intercept_
print(f"Slope (m): {m}")
print(f"Intercept (c): {c}")
```

```python
Y_pred = model.predict(X)
```

```python
Y_pred
```

```python
plt.scatter(X, Y, color='blue', label='Data points')    # Actual data points
plt.plot(X, Y_pred, color='red', label='Regression line')  # Fitted line
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.legend()
plt.title('Simple Linear Regression')
plt.show()
```
